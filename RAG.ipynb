```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.runnables import RunnableParallel
from langchain_core.output_parsers import StrOutputParser
```

```python

embedding = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001",google_api_key=os.environ['GEMINI_API_KEY'])
```

```python
vectorstore=Chroma.from_documents(documents=pages_char_split,embedding=embedding,persist_directory='/content/chroma')
```

```python
vectorstore= Chroma(persist_directory="/content/chroma",embedding_function=embedding)
```

```python
# vectorstore.
```

```python
len(vectorstore.get()['documents'])
```

```python
retriever=vectorstore.as_retriever(search_type='mmr',
                                   search_kwargs={'k':3,'lambda_mult':0.7})
```

```python
TEMPLATE="""
Answer the following question:
{question}

To answer the question, use only the following context:
{context}

At the end of the response, specify the name of the lecture this context is taken from in the format:
Resources:*Lecture Title*
where *Lecture Title should be substituted with the title of all resource lectures.
"""
```

```python
prompt_template =PromptTemplate.from_template(TEMPLATE)
```

```python
chat=ChatOpenAI(model="models/gemini-2.5-flash-lite",
                api_key=os.environ["GEMINI_API_KEY"],
              base_url="https://generativelanguage.googleapis.com/v1beta/openai/v1",
               temperature=0,


            )
```

```python
question="what software do data scientists use?"
```

```python
chain={'context':retriever,
       'question':RunnablePassthrough()}|prompt_template|chat|StrOutputParser()
```

```python
response=chain.invoke(question)
```
